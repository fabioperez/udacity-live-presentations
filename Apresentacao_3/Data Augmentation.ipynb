# Data Augmentation com Keras

## Lembretes

* Sempre visualizar os resultados da aumentação de dados para verificar se as imagens estão sendo geradas corretamente, antes de iniciar o treino.
* Verificar se os dados estão sendo passados no formato correto. Por exemplo, se o valor dos pixels vai de 0 a 255 enquanto a rede foi inicialmente treinada com valores *float* entre 0 a 1, a rede pode ter dificuldades para convergir.
* Fazer aumentações que façam sentido para o domínio do problema. Por exemplo, não faz sentido fazer um *flip* horizontal em um problema de reconhecimento de caracteres.
* As aumentações podem ser offline - quando as imagens aumentadas são geradas e armazenadas - ou online - quando as imagens são aumentadas em tempo real durante o treinamento. O Keras faz as aumentações online.
* Sempre leia a documentação!

```python
from keras.preprocessing.image import load_img, img_to_array, array_to_img
from keras.preprocessing.image import ImageDataGenerator

import skimage
from skimage.io import imshow, imread

import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [12.5, 12.5]

%load_ext autoreload
%autoreload 2
```

```python
# Função auxiliar, utilizada para exibir as imagens aumentadas.

def grid_display(gen):
    grid_image = np.concatenate([np.concatenate([next(gen)[0][0, ...].astype('uint8')
                                                for _ in range(3)], axis=1)
                                for _ in range(3)], axis=0)
    imshow(grid_image)
```

```python
def external_func_example(x):
    x = x.astype('uint8')
    x = skimage.util.random_noise(x, var=0.001)
    x *= 255.
    return x

# Documentação: https://keras.io/preprocessing/image/
# 
# Neste exemplo, estamos usando o argumento `preprocessing_function`
# para realizarmos uma aumentação que não está disponível no Keras.
# Esta função normalmente é usada para realizar algum tipo de
# preprocessamento nas imagens, mas pode ser usada também para a
# realização de aumentação de dados mais avançadas.
# Neste caso, utilizamos a adição de ruído às imagens como exemplo.

datagen = ImageDataGenerator(featurewise_center=False,
    samplewise_center=False,
    featurewise_std_normalization=False,
    samplewise_std_normalization=False,
    zca_whitening=False,
    zca_epsilon=1e-6,
    rotation_range=0.15,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.1,
    zoom_range=0.2,
    channel_shift_range=0.,
    fill_mode='reflect',
    cval=0.,
    horizontal_flip=True,
    vertical_flip=False,
    rescale=None,
    preprocessing_function=external_func_example)

examples_gen = datagen.flow_from_directory('data/sample_data/', target_size=(224, 224))
```

```python
grid_display(examples_gen)
```
