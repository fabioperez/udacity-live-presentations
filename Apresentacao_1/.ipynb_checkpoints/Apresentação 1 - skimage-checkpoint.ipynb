```python
astronaut = data.astronaut()
astr_copy = astronaut.copy()
astr_copy[:, :, 0] = 0
imshow(astr_copy)
```

```python
from skimage.io import imsave
imsave('astronaut.png', astr_copy)
```

```python
imshow(astronaut)
```

```python
from skimage.util import random_noise
from skimage.restoration import denoise_bilateral
astronaut_noise = random_noise(astronaut)
imshow(astronaut_noise)
imshow(denoise_bilateral(astronaut_noise, sigma_spatial=2))
```
